```python
from lxml import html
import requests
from bs4 import BeautifulSoup
page = requests.get('https://www.moneycontrol.com/stocks/marketstats/indexcomp.php?optex=BSE&opttopic=indexcomp&index=12')
soup = BeautifulSoup(page.text, "html.parser")
#tree = html.fromstring(page.content)
#quotes = tree.xpath('//div[@class="col-lg-12 largetable"]/text()')
x=str(soup.find('table', {'class' :'tbldata14 bdrtpg'}))
#print(quotes)
```

```python
li_s=[i for i in range(len(x)) if x.startswith('class="bl_12', i)][2:]
li_s1=[i for i in range(len(x)) if x.startswith('class="brdrgtgry', i)][8:]
```

```python
li_comp=[]
li_marketcap=[]
for i in li_s:
    li_comp.append(x[i:i+200][x[i:i+200].find('<b>')+3:x[i:i+200].find("</b>")])
li_s2=[]
for i in range(len(li_s1)):
    if i%6==0:
        li_s2.append(li_s1[i])
for i in li_s2:
    li_temp=[j for j in range(len(x[i:i+300])) if x[i:i+300].startswith('<td', j)]
    li_marketcap.append(x[i:i+300][li_temp[2]:li_temp[2]+50][36:x[i:i+300][li_temp[2]:li_temp[2]+50].find("</td>")])
```

```python
li_companies=[]
for i in range(len(li_comp)):
    if i%2==0:
        li_companies.append(li_comp[i])
for i in range(len(li_marketcap)):
    if "</t" in li_marketcap[i]:
        li_marketcap[i]=li_marketcap[i].replace("</t","")
        li_marketcap[i]=float(li_marketcap[i].replace(",",""))
```

```python
df1=pd.DataFrame(columns=["Company","MarketCap"])
df1["Company"]=li_companies
df1["MarketCap"]=li_marketcap
```

```python
df1.to_csv("bse500_marketcap.csv")
```

```python
def checkif(a,b):
    li1=[x for x in a.replace(" ","").lower()]
    li2=[x for x in b.replace(" ","").lower()]
    j=0
    k=0
    for i in range(len(li1)):
        if len(li1)>len(li2):
            break
        else:
            if li1[i]==li2[j] and j<len(li2)-1:
                k+=1
                j+=1
            elif j==len(li2)-1 and li1[i]==li2[j]:
                k+=1
                break
            elif li1[i]!=li2[j]:
                j+=1
    if k==len(li1):
        return True
    else:
        return False
for i in range(len(li)):
    for j in range(len(li2)):
        if checkif(li2[j],li[i]):
            li_1.append(li1[j])
            li_2.append(li2[j])
            li_3.append(li3[j])
```
